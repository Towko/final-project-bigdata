# Завдання 5. Визначення популярності Python

Інструмент, який потрібен для виконання цього завдання: Pandas (робота з текстовими даними, підрахунок відсотків)

Опис завдання: Обчисли відсоток респондентів, які програмують на Python. Зверни увагу, що поле з мовами програмування може містити множинні значення, тому потрібно перевірити наявність Python в цьому полі.

Фінальний результат: Відсоток респондентів, які програмують на Python (наприклад, 45.3%).

Формат здачі: Комірка з кодом та виведеним відсотковим значенням.

```python
# Створюємо новий стовпчик worked_with_python
df_data["worked_with_python"] = False
```

```python
# Кількість респондентів, які програмують на Python
df_data["worked_with_python"] = df_data["LanguageHaveWorkedWith"].str.contains("Python", case=False, na=False)
py_count = df_data["worked_with_python"].sum()

# Загальна кількість респондентів
total_sum = df_data["LanguageHaveWorkedWith"].notna().sum()

# Обчислюємо відсоток респондентів, які програмують на Python
py_percent = py_count / total_sum * 100

print("Кількість респондентів, які програмують на Python:", py_count)
print(f"Відсоток: {py_percent:.2f}%")
```

# Завдання 7. Географічний аналіз компенсації Python-розробників

Інструмент, який потрібен для виконання цього завдання: Pandas (групування даних, агрегація, фільтрація)

Опис завдання: Серед респондентів, що програмують на Python, проведи аналіз компенсації (ConvertedCompYearly) в розрізі країн. Для кожної країни обчисли:

Середню суму компенсації
Медіанну суму компенсації

Фінальний результат: DataFrame з колонками: країна, середня компенсація, медіанна компенсація.

Формат здачі: Комірка з кодом та виведеною таблицею результатів.

```python
# Фільтруємо тих хто працював в Python
py_worked = df_data.worked_with_python == True
```

```python
# Видаляємо рядки де відсутнє значення зарплати
py_filtered = df_data[py_worked & df_data["ConvertedCompYearly"].notna()]
```

```python
# Групуємо дані за країною в нову таблицю
country_python = (
    py_filtered
    .groupby("Country")["ConvertedCompYearly"]
    .agg(mean_salary="mean", median_salary="median")
    .round(2)
    .reset_index()
    .rename(columns={"Country": "Країна", "mean_salary":"Середня компенсація", "median_salary":"Медіанна компенсація"})
)
display(country_python)
```

# Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів

Інструмент, який потрібен для виконання цього завдання: Pandas (сортування даних, вибірка топ-записів)

Опис завдання: Визнач рівні освіти 5 респондентів з найбільшою компенсацією. Тобі потрібно:

Відсортувати респондентів за рівнем компенсації
Вибрати топ-5
Показати їх рівні освіти
На що ментор зверне увагу при перевірці роботи:

Чи правильно відсортовано дані за компенсацією (у спадному порядку)
Чи вибрано саме 5 респондентів
Чи відображено їх рівні освіти
Чи враховані можливі пропущені значення
Фінальний результат: Список або таблиця з 5 рівнями освіти найбільш оплачуваних респондентів.

Формат здачі: Комірка з кодом та виведеним списком або таблицею рівнів освіти.

```python
# Групуємо дані в нову таблицю з новими колонками
education = (
    py_filtered
    .sort_values("ConvertedCompYearly", ascending=False)
    .loc[:,["ResponseId", "EdLevel", "ConvertedCompYearly"]]
    .head(5)
    .reset_index(drop=True)
)
education.columns = ["ResponseId", "Рівень освіти", "Компенсація"]

display(education)
```

Інструмент, який потрібен для виконання цього завдання: Pandas (групування, підрахунок відсотків, агрегація)

Опис завдання: Для кожної вікової категорії визнач відсоток респондентів, які програмують на Python. Це дасть уявлення про те, в яких вікових групах Python є найпопулярнішим.

Фінальний результат: Таблиця з колонками: вікова категорія, відсоток Python-розробників.

Формат здачі: Комірка з кодом та виведеною таблицею результатів.

```python
# Групуємо дані по колонці worked_with_python та рахуємо відсоток респондентів та скільки працюють з Python
popularity = (
    df_data
    .groupby("Age")["worked_with_python"]
    .agg(total="count", python="sum")
    .assign(percent=lambda x: (x["python"] / x["total"]) * 100)
    .round(2)
)

display(popularity)
```

# *Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників
